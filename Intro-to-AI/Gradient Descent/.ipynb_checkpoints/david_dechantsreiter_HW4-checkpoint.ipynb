```python

# cleaning the data, especially handling the \t occurance
df['pcv'] = df['pcv'].str.replace('\t', '')
df['wbcc'] = df['wbcc'].str.replace('\t', '')
df['rbcc'] = df['wbcc'].str.replace('\t', '')

# replacing missing values with NaNs
df['dm'] = df['dm'].str.strip()
df['dm'] = df['dm'].replace('', np.nan)

df['cad'] = df['cad'].str.strip()
df['cad'] = df['cad'].replace('', np.nan)

df['class'] = df['class'].str.strip()

df.replace('?', pd.NA, inplace=True)


# Converting to correct datatypes
df['age'] = pd.to_numeric(df['age'])
df['bp'] = pd.to_numeric(df['bp'])
df['sg'] = pd.to_numeric(df['sg'])
df['al'] = pd.to_numeric(df['al'])
df['su'] = pd.to_numeric(df['su'])
df['bgr'] = pd.to_numeric(df['bgr'])
df['bu'] = pd.to_numeric(df['bu'])
df['sc'] = pd.to_numeric(df['sc'])
df['sod'] = pd.to_numeric(df['sod'])
df['pot'] = pd.to_numeric(df['pot'])
df['hemo'] = pd.to_numeric(df['hemo'])
df['pcv'] = pd.to_numeric(df['pcv'])
df['wbcc'] = pd.to_numeric(df['wbcc'])
df['rbcc'] = pd.to_numeric(df['rbcc'])
```

```python
# This function replaces each missing value with mean*

# mean* refers to grouped mean of the target class value, this is done to add some more information to the
# data instead of just replacing with overall mean

def replace_na_meanclass(df, col):
    
    df[df[col].isna()]
    
    cols = ['class_ckd']
    
    filtered_rows_1 = df[(df[['class']] == 'ckd').all(axis=1)]
    filtered_rows_0 = df[(df[['class']] == 'notckd').all(axis=1)]
    
    mean_1 = filtered_rows_1[col].mean()
    mean_0 = filtered_rows_0[col].mean()
    
    # Replace NaN values in 'col' based on class_ckd
    df.loc[df['class'] == 'ckd', col] = df.loc[df['class'] == 'ckd', col].fillna(mean_1)
    df.loc[df['class'] == 'notckd', col] = df.loc[df['class'] == 'notckd', col].fillna(mean_0)
```

```python
# Replacing for numerical features
replace_na_meanclass(df, 'age')
replace_na_meanclass(df, 'bp')
replace_na_meanclass(df, 'sg')
replace_na_meanclass(df, 'al')
replace_na_meanclass(df, 'su')
replace_na_meanclass(df, 'bgr')
replace_na_meanclass(df, 'bu')
replace_na_meanclass(df, 'sc')
replace_na_meanclass(df, 'sod')
replace_na_meanclass(df, 'pot')
replace_na_meanclass(df, 'hemo')
replace_na_meanclass(df, 'pcv')
replace_na_meanclass(df, 'wbcc')
replace_na_meanclass(df, 'rbcc')


# Replacing for binary features (just replacing with mode)
df['rbc'].fillna(df['rbc'].mode()[0], inplace=True)
df['pc'].fillna(df['pc'].mode()[0], inplace=True)
df['pcc'].fillna(df['pcc'].mode()[0], inplace=True)
df['ba'].fillna(df['ba'].mode()[0], inplace=True)
df['htn'].fillna(df['htn'].mode()[0], inplace=True)
df['dm'].fillna(df['dm'].mode()[0], inplace=True)
df['cad'].fillna(df['cad'].mode()[0], inplace=True)
df['appet'].fillna(df['appet'].mode()[0], inplace=True)
df['pe'].fillna(df['pe'].mode()[0], inplace=True)
df['ane'].fillna(df['ane'].mode()[0], inplace=True)
```

# Gradient Descent Implementation

```python
import numpy as np

def sigmoid(z):
    z = np.clip(z, -500, 500)  # Prevent numerical overflow
    return 1 / (1 + np.exp(-z))


#Calculates loss for a classification task
def logistic_cost(X, y, w, reg_lambda):
    m = len(y)
    
    predictions = sigmoid((np.dot(X, w)))
    
    log_const = 1e-4
    
    cost = -(1/m) * np.sum(y * np.log(predictions + log_const) + (1-y) * np.log(1 - predictions + log_const))
    
    regularization_term = (reg_lambda/(2*m)) * np.sum(np.square(w[1:]))
    
    return cost + regularization_term


# Calculates Gradient of logistic_cost function
def logistic_gradient(w, X, y, reg_lambda):
    m = len(y)
    
    predictions = sigmoid(np.dot(X, w))
    
    reg_term = (reg_lambda / m) * np.concatenate([[0], w[1:]])
    
    grad = (1 / m) * np.dot(X.T, (predictions - y)) + reg_term
    
    return grad

#I Implementes gradient descent
def gradient_descent_logistic(X, y, learning_rate, iterations, reg_lambda, tolerance=1e-4):
    
    X = np.c_[np.ones(X.shape[0]), X]  # Add bias term to input features
    
    m, n = X.shape
    
    w = np.random.randn(X.shape[1]) * 0.01  # Initialize weights
    
    cost_history = []
    
    w_old = np.random.randn(X.shape[1])

    for i in range(iterations):
        grad = logistic_gradient(w, X, y, reg_lambda)
        w -= learning_rate * grad
        
        cost = logistic_cost(X, y, w, reg_lambda)
        cost_history.append(cost)
        
        # Check for convergence
        if np.linalg.norm(w - w_old) <= tolerance:
            print(f'Converged after {i + 1} iterations. Norm: {np.linalg.norm(w - w_old):.6f}')
            return w
        
        w_old = w.copy()

    print('Did not converge!')
    return w
```

```python
weights = gradient_descent_logistic(X_train, y_train, learning_rate = 0.01, iterations = 10000, reg_lambda = -0.5)
```

```python
def predict(X, w):
    X = np.c_[np.ones(X.shape[0]), X]  # Add bias term to input features
    return (sigmoid(np.dot(X, w)) >= 0.5).astype(int)
```

```python
# testing algorithm with unstandardized data

weights = gradient_descent_logistic(X_train, y_train, learning_rate = 0.01, iterations = 1000, reg_lambda = 0.3)
y_pred = predict(X_test, weights)
print(y_pred)
```

```python
def f1_measure(y_pred, y_real):
    TP = 0
    TN = 0
    FP = 0
    FN = 0
    
    for i in range(len(y_real)):
        if y_real.iloc[i] == 1 and y_pred[i] == 1:
            TP += 1
        elif y_real.iloc[i] == 0 and y_pred[i] == 0:
            TN += 1
        elif y_real.iloc[i] == 0 and y_pred[i] == 1:
            FP += 1
        else:
            FN += 1
            
    # Handling division by zero
    Pre = TP / (TP + FP) if (TP + FP) > 0 else 0
    Rec = TP / (TP + FN) if (TP + FN) > 0 else 0
    F1 = 2 * (Pre * Rec) / (Pre + Rec) if (Pre + Rec) > 0 else 0
    
    return F1
```

```python
print(f1_measure(y_pred, y_test))
```

```python
f1_measures = []

for lamb in [-2, 0.00001, 4]:
    
    weights = gradient_descent_logistic(X_train, y_train, learning_rate = 0.01, iterations = 1000, reg_lambda = lamb)
    y_pred = predict(X_test, weights)
    f1_measures.append(f1_measure(y_pred, y_test))
```

# Plotting F1 Measure for Unstandardized and Standardized Data

```python
import matplotlib.pyplot as plt

plt.plot([-2, 0.00001, 4], f1_measures)
plt.title("Plotting the F1-Measure for Range of Regularization Values (unstandardized)")
plt.xlabel('Lambda')
plt.ylabel('F1 Measure')
plt.show()
```

# Data Standardization

```python
df_standardized = df_encoded.copy()
```

```python
for col in df_standardized:
    
    if df_standardized[col].nunique() > 2:
        
        mu = np.mean(df_standardized[col])
        
        std = np.std(df_standardized[col])
        
        df_standardized[col] = (df_standardized[col] - mu) / std
```

```python
from sklearn.model_selection import train_test_split

X_norm = df_standardized.iloc[:, :-1]
y_norm = df_standardized.iloc[:, -1]

X_train_norm, X_test_norm, y_train_norm, y_test_norm = train_test_split(X_norm, y_norm, test_size = 0.2)
```

```python
weights = gradient_descent_logistic(X_train_norm, y_train_norm, learning_rate = 0.01, iterations = 1000, reg_lambda = 0.3)
y_pred = predict(X_test_norm, weights)
print(y_pred)
print(f1_measure(y_pred, y_test_norm))
```

```python
f1_measures_norm_test = []
f1_measures_norm_train = []

for lamb in np.arange(-2, 4, 0.2):
    print('Current Lambda: ', lamb)
    weights = gradient_descent_logistic(X_train_norm, y_train_norm, learning_rate = 0.01, iterations = 1000, reg_lambda = lamb)
    y_pred_test = predict(X_test_norm, weights)
    y_pred_train = predict(X_train_norm, weights)
    f1_measures_norm_test.append(f1_measure(y_pred_test, y_test_norm))
    f1_measures_norm_train.append(f1_measure(y_pred_train, y_train_norm))
    print('\n')
print('Complete')
```

```python
import matplotlib.pyplot as plt

plt.plot(np.arange(-2, 4, 0.2), f1_measures_norm_train, label = 'train')
plt.plot(np.arange(-2, 4, 0.2), f1_measures_norm_test, label = 'test')
plt.legend()
plt.title("Plotting the F1-Measure for Range of Regularization Values (standardized)")
plt.xlabel('Lambda')
plt.ylabel('F1 Measure')
plt.show()
```
